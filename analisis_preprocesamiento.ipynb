```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
df = pd.read_csv('/Users/enriquepuenteruiz/Downloads/archive/Chicago_Crimes_2012_to_2017.csv')
```

```python
# AnÃ¡lisis de distribuciÃ³n de algunas variables importantes
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Primary Type', order=df['Primary Type'].value_counts().index)
plt.xticks(rotation=90)
plt.title('DistribuciÃ³n de Tipos de Crimen')
plt.xlabel('Tipo de Crimen')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Year', order=df['Year'].value_counts().index)
plt.title('DistribuciÃ³n de CrÃ­menes por AÃ±o')
plt.xlabel('AÃ±o')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Mapa con coordenadas (latitud y longitud)
crime_map = folium.Map(location=[41.8781, -87.6298], zoom_start=10)  # UbicaciÃ³n centrada en Chicago

# AÃ±adir puntos al mapa (muestra aleatoria de 500 puntos para no sobrecargar el mapa)
df_sample = df.dropna(subset=['Latitude', 'Longitude']).sample(n=500)
for _, row in df_sample.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=3,
        color='red',
        fill=True,
        fill_opacity=0.5
    ).add_to(crime_map)
```

```python
# Guardar el mapa como archivo HTML
crime_map.save('chicago_crimes_map.html')
print("Mapa guardado como 'chicago_crimes_map.html'")
```
